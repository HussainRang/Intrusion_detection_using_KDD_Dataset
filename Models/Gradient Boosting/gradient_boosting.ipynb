### TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split

X = data.drop(['attack_category'],axis=1)
y = data['attack_category']

X_train,X_test,y_train,y_test = train_test_split( X , y , test_size=0.3 , random_state=42 , stratify = y )
```

```python
X_train.shape , y_train.shape
```

```python
X_test.shape , y_test.shape
```

<br/>
<br/>
<br/>

#### REINDEXING DATA

```python
X_train = X_train.reset_index(drop = True)
y_train = y_train.reset_index(drop = True)
```

```python
X_train
```

```python
y_train
```

```python
X_test = X_test.reset_index(drop = True)
y_test = y_test.reset_index(drop = True)
```

```python
X_test
```

```python
y_test
```

<br/>
<br/>
<br/>

### MODEL

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_classifier = GradientBoostingClassifier()
gb_classifier.fit(X_train,y_train)
```

```python
y_pred = gb_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
y_true = y_test
target_names = [ 'dos', 'normal', 'probe' , 'r2l' , 'u2r' ]
report = classification_report(y_true, y_pred, target_names=target_names)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_true,y_pred)
pd.DataFrame(confusion_matrix , index = target_names , columns = target_names)
```

## APPLYING GRID SEARCH CV

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV( GradientBoostingClassifier( random_state=42 , min_samples_leaf=8 ) , {
    'learning_rate' : [0.05,0.1],
    'n_estimators' : [100,200],
    'max_leaf_nodes' : [24,32],
} , cv=5 ,return_train_score = False , verbose=5 )

clf.fit(X,y)
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df = df[['param_n_estimators' , 'param_max_leaf_nodes' , 'param_learning_rate' , 'mean_test_score']]
df.sort_values(by=['mean_test_score'] , ascending=False)
```

```python
gb_classifier = GradientBoostingClassifier(random_state=42 , n_estimators=200 , max_leaf_nodes=24 , learning_rate=0.05 )
gb_classifier.fit(X_train,y_train)
y_pred = gb_classifier.predict(X_test)
```

```python
y_true = y_test
target_names = [ 'dos', 'normal', 'probe' , 'r2l' , 'u2r' ]
report = classification_report(y_true, y_pred, target_names=target_names)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_true,y_pred)
pd.DataFrame(confusion_matrix , index = target_names , columns = target_names)
```

```python
from sklearn.metrics import accuracy_score , precision_score , recall_score , f1_score 
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score( y_true , y_pred , average = None )
recall = recall_score( y_true , y_pred , average = None )
f1score = f1_score( y_true , y_pred , average = None )
```

```python
print(accuracy)
print(precision)
print(recall)
print(f1score)
```
