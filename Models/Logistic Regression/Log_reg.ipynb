```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split

X = df.drop(['attack_category'],axis=1)
y = df['attack_category']

X_train,X_test,y_train,y_test = train_test_split( X , y , test_size=0.3 , random_state=42 , stratify = y)
```

```python
X_train.columns
```

```python
y_train.head(5)
```

<br/>
<br/>

#### REINDEXING DF

```python
X_train = X_train.reset_index(drop = True)
y_train = y_train.reset_index(drop = True)
```

```python
X_train
```

```python
y_train
```

```python
X_test = X_test.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

```python
X_test
```

```python
y_test
```

<br/>
<br/>

### MODEL

```python
from sklearn.linear_model import LogisticRegression

# multinomial set for softmax regression
model = LogisticRegression(solver='lbfgs', multi_class='multinomial')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
y_true = y_test
target_names = [ 'dos', 'normal', 'probe' , 'r2l' , 'u2r' ]
report = classification_report(y_true, y_pred, target_names=target_names)
```

```python
print(report)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_true,y_pred)
print(confusion_matrix)
```

```python
pd.DataFrame(confusion_matrix , index = target_names , columns = target_names)
```

```python
from sklearn.metrics import accuracy_score , precision_score , recall_score , f1_score 
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score( y_true , y_pred , average = None )
recall = recall_score( y_true , y_pred , average = None )
f1score = f1_score( y_true , y_pred , average = None )
```

```python
print(accuracy)
print(precision)
print(recall)
print(f1score)
```

```python
data = {
    "model" : "Logistic Regression",
    "solver" : "lbfgs",
    "multiclass" : "multinomial",
    "accuracy" : accuracy ,
    "precision" : precision.tolist() ,
    "recall" : recall.tolist() ,
    "F1_score" : f1score.tolist() ,
    "class_names" : target_names
}

data
```
