```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../../data/preprocessed_data.csv')
data.head(5)
```

```python
X = data.drop(['attack_category'],axis=1)
y = data['attack_category']
```

<br/>
<br/>
<br/>

### TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split( X , y , test_size = 0.3 , random_state = 42 , stratify = y )
```

```python
X_train.columns , X_train.shape
```

```python
X_test.shape
```

<br/>
<br/>

#### REINDEXING DATAFRAME

```python
X_train = X_train.reset_index(drop = True)
y_train = y_train.reset_index(drop = True)

X_test = X_test.reset_index(drop = True)
y_test = y_test.reset_index(drop = True)
```

```python
X_train.head(5)
```

```python
X_test.head(5)
```

```python
y_train.head(5)
```

```python
y_test.head(5)
```

<br/>
<br/>
<br/>

### MODEL

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier()
```

```python
rf_classifier.fit(X_train,y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
y_true = y_test
target_names = [ 'dos', 'normal', 'probe' , 'r2l' , 'u2r' ]
report = classification_report(y_true, y_pred, target_names=target_names)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_true,y_pred)
pd.DataFrame(confusion_matrix , index = target_names , columns = target_names)
```

<br/>
<br/>
<br/>

### APPLYING GRID SEARCH

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV( RandomForestClassifier( random_state = 42 ) , {
    'n_estimators':[200,300,],
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'max_depth' : [10,15,20],
    'bootstrap' : [True,False]
} , cv=5 ,return_train_score = False , verbose = 5)

clf.fit(X,y)
```

```python
df = pd.DataFrame(clf.cv_results_)
df = df[['param_n_estimators' , 'param_criterion' , 'param_max_depth', 'param_bootstrap' , 'mean_test_score']]
df.sort_values(by=['mean_test_score'] , ascending=False)
```

## FINAL MODEL

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(
                                        random_state = 42 , n_estimators = 300 ,
                                        criterion='log_loss' , max_depth=20 ,
                                        bootstrap=False 
                                      )
```

```python
rf_classifier.fit(X_train,y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
y_true = y_test
target_names = [ 'dos', 'normal', 'probe' , 'r2l' , 'u2r' ]
report = classification_report(y_true, y_pred, target_names=target_names)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_true,y_pred)
pd.DataFrame(confusion_matrix , index = target_names , columns = target_names)
```

```python
from sklearn.metrics import accuracy_score , precision_score , recall_score , f1_score 
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score( y_true , y_pred , average = None )
recall = recall_score( y_true , y_pred , average = None )
f1score = f1_score( y_true , y_pred , average = None )


print(accuracy)
print(precision)
print(recall)
print(f1score)
```

```python
data = {
    "model" : "Logistic Regression",
    "n_estimators": 300 ,
    "criterion": 'log_loss' ,
    "max_depth": 20 ,
    "bootstrap": False ,
    "accuracy" : accuracy ,
    "precision" : precision.tolist() ,
    "recall" : recall.tolist() ,
    "F1_score" : f1score.tolist() ,
    "class_names" : target_names
}

data
```
