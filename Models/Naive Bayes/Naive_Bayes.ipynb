```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../../data/preprocessed_data.csv')
data.head(5)
```

```python
X = data.drop(['attack_category'],axis=1)
y = data['attack_category']
```

```python
X.columns
```

<br/>
<br/>
<br/>

### TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split( X , y , test_size = 0.3 , random_state = 42 , stratify = y )
```

```python
X_train.shape , X_test.shape
```

```python
y_train.shape , y_test.shape
```

<br/>
<br/>

#### REINDEXING DATAFRAME

```python
X_train = X_train.reset_index( drop = True )
y_train = y_train.reset_index( drop = True )

X_test = X_test.reset_index( drop = True )
y_test = y_test.reset_index( drop = True )
```

```python
X_train
```

```python
X_test
```

<br/>
<br/>
<br/>

### MODEL

```python
from sklearn.naive_bayes import GaussianNB , BernoulliNB

gnb = GaussianNB()
bnb = BernoulliNB()
```

```python
gnb.fit(X_train,y_train)
y_pred_gnb = gnb.predict(X_test)
```

```python
bnb.fit(X_train,y_train)
y_pred_bnb = bnb.predict(X_test)
```

```python
from sklearn.metrics import classification_report
y_true = y_test
target_names = [ 'dos', 'normal', 'probe' , 'r2l' , 'u2r' ]

print("=====================GAUSSIAN NAIVE BAYES====================")
report = classification_report(y_true, y_pred_gnb, target_names=target_names)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_true,y_pred_gnb)
pd.DataFrame(conf_mat , index = target_names , columns = target_names)
```

```python
print("================BERNOULLI NAIVE BAYES================")
report = classification_report(y_true, y_pred_bnb, target_names=target_names)
print(report)
```

```python
conf_mat = confusion_matrix( y_true , y_pred_bnb )
pd.DataFrame(conf_mat , index = target_names , columns = target_names)
```

<br/>
<br/>
<br/>

#### SAVING DATA FOR GAUSSIAN NAIVE BAYES

```python
from sklearn.metrics import accuracy_score , precision_score , recall_score , f1_score 

accuracy = accuracy_score(y_true, y_pred_gnb)
precision = precision_score( y_true , y_pred_gnb , average = None )
recall = recall_score( y_true , y_pred_gnb , average = None )
f1score = f1_score( y_true , y_pred_gnb , average = None )

print(accuracy)
print(precision)
print(recall)
print(f1score)
```

```python
data = {
    "model" : "Gaussian Naive Bayes",
    "accuracy" : accuracy ,
    "precision" : precision.tolist() ,
    "recall" : recall.tolist() ,
    "F1_score" : f1score.tolist() ,
    "class_names" : target_names
}

data
```

```python
import json
with open("../../model_performances/gaussian_naive_bayes.json", "w") as outfile:
    json.dump(data, outfile)
```
