#### DATA PREPERATION

- Dropping unwanted coloumn
- Splitting in X and Y
- OHE and Standardization
- Label Encoding
- Merging the Dataset

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

def prepare_dataset(df:pd.DataFrame):
    
    # Dropping unwanted column
    df.drop(['Unnamed: 0'],axis=1,inplace=True)

    # Splitting in X and Y
    X = df.drop(['attack_category'],axis=1)
    Y = df['attack_category']

    # Splitting categorical column and numerical column
    categorical_features = X.select_dtypes(include=['object','category'])
    numerical_features = X.select_dtypes(exclude=['object','category'])

    # OHE categorical coloumns
    ohe_categorical = pd.get_dummies(categorical_features,drop_first=True,prefix=categorical_features.columns)
    
    # Standardizing numerical coloumn
    standard_scaler = StandardScaler()
    cols_list = numerical_features.columns
    standard_scaler.fit(numerical_features)
    
    numerical_features = standard_scaler.transform(numerical_features)
    numerical_features = pd.DataFrame(numerical_features,columns=cols_list)
    
    # Label Encoding target column
    label_encoder = LabelEncoder()
    label_encoder.fit(Y)
    Y_new = label_encoder.transform(Y)
    
    # Merging all the training coloumns
    X_new = pd.merge(numerical_features, ohe_categorical, left_index=True, right_index=True)  

    return X_new , Y_new
```
