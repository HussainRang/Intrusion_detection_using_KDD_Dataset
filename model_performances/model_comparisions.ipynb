```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd
```

```python
file_names = os.listdir()
if '.ipynb_checkpoints' in file_names:
    file_names.remove('.ipynb_checkpoints')
if 'model_comparisions.ipynb' in file_names:
    file_names.remove('model_comparisions.ipynb')
```

```python
file_names
```

```python
import json
performance_data = []

for i in file_names:
    f = open(i)
    data = json.load(f)
    performance_data.append(data)
    f.close()
```

# PRECISION VS MODELS

```python
prec_dos = []
prec_normal = []
prec_probe = []
prec_r2l = []
prec_u2r = []
models=[]


for i in performance_data:  
    prec_dos.append(i['precision'][0])
    prec_normal.append(i['precision'][1])
    prec_probe.append(i['precision'][2])
    prec_r2l.append(i['precision'][3])
    prec_u2r.append(i['precision'][4])
    models.append(i['model'])
    
print(len(prec_dos))
```

```python
plt.figure(figsize=(20,20))

br1 = np.arange(len(prec_dos))
br1 = br1*5
br2 = [i+0.80 for i in br1]
br3 = [i+0.80 for i in br2]
br4 = [i+0.80 for i in br3]
br5 = [i+0.80 for i in br4]

fig, ax = plt.subplots()
ax.barh(y=br1 , width=prec_dos , label = 'DOS')
ax.barh(y=br2 , width=prec_normal , label = 'Normal')
ax.barh(y=br3 , width=prec_probe , label = 'Probe')
ax.barh(y=br4 , width=prec_r2l , label = 'r2l')
ax.barh(y=br5 , width=prec_u2r , label='u2r')

plt.ylabel('MODELS', fontweight ='bold', fontsize = 15)
plt.xlabel('PRECISION', fontweight ='bold', fontsize = 15)
plt.yticks(br1,models)
plt.legend(bbox_to_anchor=(1.23,1))
plt.show();
```

# ACCURACY VS MODELS

```python
accuracy = []
models = []

for i in performance_data:
    accuracy.append(i['accuracy'])
    models.append(i['model'])
accuracy
```

```python
plt.figure(figsize=(15,15))

br1 = np.arange(len(accuracy))
br1 = br1*1

fig, ax = plt.subplots()
ax.barh(y=br1 , width=accuracy)

plt.ylabel('MODELS', fontweight ='bold', fontsize = 15)
plt.xlabel('ACCURACY', fontweight ='bold', fontsize = 15)
plt.yticks(br1,models)
plt.legend(bbox_to_anchor=(1.23,1))
plt.show();
```

```python
df = pd.DataFrame({ "Model Name": models , "Accuracy" : accuracy })
df = df.sort_values(by=['Accuracy'] , ascending=False)
df = df.reset_index(drop=True)
df
```
